```python
kgroup = 2
x = newdata.groupby(by=['Label']).size().reset_index(name='Count').sort_values(['Count'], ascending=True,ignore_index=True)

for i in range(0,len(x)-kgroup):
    newdata.drop(newdata.index[newdata['Label'] == x['Label'][i]], inplace=True)

x = newdata.groupby(by=['Label']).size().reset_index(name='Count').sort_values(['Label'], ascending=True,ignore_index=True)

newdata.sort_values(by=['Label'],inplace=True,ignore_index=True)

MyColumns=['Open','High','Low','Close']
for n in range(0,m-1):
    for w in range(0,4):
        MyColumns.append(MyColumns[w]+str(n+1))
MyColumns.append('Label')     

df=pd.DataFrame(columns=MyColumns)

for i in range(0,kgroup):
    df = df.append(newdata[0:min(x['Count'])],ignore_index=True)
    newdata.drop(newdata.index[newdata['Label'] == x['Label'][i]], inplace=True)
    newdata.sort_values(by=['Label'],inplace=True,ignore_index=True)
    df['Label'][df.index.values[df['Label'] == x['Label'][i]]]=i
    
print(df.to_string())    
```

```python
from sklearn.model_selection import train_test_split
import sklearn as sk

x_train, x_test, y_train, y_test = train_test_split(df[MyColumns[:-1]],df[MyColumns[-1]] ,test_size = 0.2) 
x_train = sk.preprocessing.normalize(x_train, norm='max', axis=1)
x_test = sk.preprocessing.normalize(x_test, norm='max', axis=1)
X_train = x_train.reshape(-1,10,4)
X_test = x_test.reshape(-1,10,4)
Y_train = np.array(y_train)
Y_test = np.array(y_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization

model = Sequential()
model.add(LSTM(128, input_shape=(X_train.shape[1:]), return_sequences=True))
model.add(Dropout(0.3))
model.add(BatchNormalization())
model.add(LSTM(64, return_sequences=True))
model.add(Dropout(0.3))
model.add(BatchNormalization())
model.add(LSTM(32))
model.add(Dropout(0.3))
model.add(BatchNormalization())
model.add(Dense(32, activation='relu'))
model.add(Dense(2, activation='softmax'))


model.compile(loss='sparse_categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])

history = model.fit(X_train, Y_train, batch_size=50, epochs=50, validation_data=(X_test, Y_test))
```
